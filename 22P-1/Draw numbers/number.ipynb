```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

### Создаем класс для наших чисел

```python
class_names = ['Ноль', "Один", "Два", "Три",
              "Четыри", "Пять", "Шесть", "Семь", "Восемь",
              "Девять"]
```

# Предобработка данных

```python
x_train.shape, x_test.shape
```

### Делим на основные и тренировочные данные

```python
y_train.shape, y_test.shape
```

### Выводим результат того, что у нас есть

```python
plt.figure()
plt.imshow(x_train[4])
plt.colorbar()
plt.grid(False)
plt.show()
```

```python
x_train = x_train / 255
x_test = x_test / 255
```

```python
plt.figure()
plt.imshow(x_train[0])
plt.colorbar()
plt.grid(False)
plt.show()
```

### Отобразим все числа, которые у нас имеются

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5, 10, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(x_train[i])
    plt.xlabel(class_names[y_train[i]])
```

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5, 5, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(x_train[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[y_train[i]])
```

```python
model = keras.Sequential([keras.layers.Flatten(input_shape=(28,28)),
                         keras.layers.Dense(128, activation='relu'),
                         keras.layers.Dense(10, activation='softmax')])
```

```python
model.compile(optimizer=tf.keras.optimizers.SGD(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

### Обучение 

```python
model.fit(x_train, y_train, epochs=300)
```

# Проверяем точность модели

```python
test_loss, test_acc = model.evaluate(x_test, y_test)
print('Test accuracy:', test_acc)
```
