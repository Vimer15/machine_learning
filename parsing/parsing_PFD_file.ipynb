```python
def extract_text_from_pdf(pdf):
    doc = fitz.open(pdf)
    print("–ù–∞–∑–≤–∞–Ω–∏–µ: ", doc.name)
    text = []
    for current_page in range(len(doc)):
        page = doc.load_page(current_page)
        page_text = page.get_text("text")
        text.append(page_text)
    return text
```

```python
def pars_pdf(text):
    date_publish = "N/A"
    for page in text:
        page_text = page.split('\n')
        page_text = [line for line in page_text if not re.fullmatch(r', ', line)]
        if '–†–µ–π—Ç–∏–Ω–≥' in page_text:
            num = page_text.index('–†–µ–π—Ç–∏–Ω–≥')
            title = page_text[num+6]
            name = page_text[num+1]
            rating = page_text[num-1]
            description = page_text[num+2]
            category = page_text[num+1]
            for index, item in enumerate(page_text):
                if "–Ω–∞–∑–∞–¥" in item:
                    date_publish = page_text[index]
                    print(date_publish)
                    if not date_publish.strip()[0].isdigit():
                        date_publish = " ".join(date_publish.split()[1:]).lstrip()
                        print(date_publish)
            content = " ".join(" ".join(text).split("\n")).replace(rating, "").replace(name, "").replace(title, "").replace(description, "").replace(date_publish, "").replace("–†–µ–π—Ç–∏–Ω–≥", "", 1)
            if(date_publish != None):
                break
             
    return  title, name, description, rating, category, date_publish, content
```

```python
# –ü–µ—Ä–µ–±–æ—Ä —Ç–µ–∫—Å—Ç–∞ –∏–∑ pdf —Ñ–∞–π–ª–∞, –ø–∞—Ä—Å–∏–Ω–≥
Brak = []
Title = []
NameCompany = []
Description = []
Raiting = []
Category = []
DataPublish = []
TextArticle = []
for pdf in all_pdf:
    try:  
        Text = extract_text_from_pdf(pdf)
        t, n, desc, r, cat, date_publish, text = pars_pdf(Text)
        Title.append(t)
        NameCompany.append(n)
        Description.append(desc)
        Raiting.append(r)
        Category.append(cat)
        DataPublish.append(date_publish)
        TextArticle.append(text)
    except Exception as e:
        print(e)
        Brak.append(pdf)
```

```python
# –°–æ–∑–¥–∞–Ω–∏–µ DataFrame
df = pd.DataFrame({
'–ó–∞–≥–æ–ª–æ–≤–æ–∫': Title,
'–ù–∞–∑–≤–∞–Ω–∏–µ –∫–æ–º–ø–∞–Ω–∏–∏': NameCompany,
'–û–ø–∏—Å–∞–Ω–∏–µ': Description,
'–†–µ–π—Ç–∏–Ω–≥': Raiting,
'–ö–∞—Ç–µ–≥–æ—Ä–∏—è': Category,
'–î–∞—Ç–∞': DataPublish,
'–¢–µ–∫—Å—Ç': TextArticle
})
```

```python
def remove_rate(text):
    words = text.split()
    if '–û—Ü–µ–Ω–∫–∞' in words:
        index = words.index('–û—Ü–µ–Ω–∫–∞')
        return ' '.join(words[index + 1:])
    return ' '.join(words)

def remove_average(text):
    match = re.search(r"([–ê-–Ø–∞-—èA-Za-z]+\d+)", text)
    if match:
        matched_word = match.group(0)
        words = text.split()
        if matched_word in words:
            i = words.index(matched_word)
            return ' '.join(words[i + 2:])
    return text

def remove_autor(text):
    words = text.split()
    if '–ê–í–¢–û–†–û–ú' in words:
        for word in ["–ö–ê–ö", "–°–¢–ê–¢–¨", "–ê–í–¢–û–†–û–ú"]:
            while word in words:
                words.remove(word)
    for company in ["redmadrobot", "artyomsoft", "TourmalineCore"]:
        if text.startswith(company):
            words = words[len(company) + 1:]
    return ' '.join(words)
```

```python
df['–¢–µ–∫—Å—Ç'] = df['–¢–µ–∫—Å—Ç'].apply(remove_rate)
df['–¢–µ–∫—Å—Ç'] = df['–¢–µ–∫—Å—Ç'].apply(remove_average)
df['–¢–µ–∫—Å—Ç'] = df['–¢–µ–∫—Å—Ç'].apply(remove_autor)
```
